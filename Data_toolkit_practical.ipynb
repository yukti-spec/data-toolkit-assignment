Question 2.  Write a Pandas script to find the mean of a specific column in a DataFrame.

Ans:  

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)
age_mean = df['Age'].mean()

print("DataFrame:")
print(df)
print("\nMean of the 'Age' column:")
print(age_mean)
```

Question 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

Ans:  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [1, 3, 2, 5, 4]
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Question 5. Generate a bar plot using Plotly.



```python
import plotly.graph_objects as go
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 14, 23, 17]
fig = go.Figure(data=[
    go.Bar(x=categories, y=values, marker_color='blue')
])
fig.update_layout(
    title='Bar Plot Example',
    xaxis_title='Categories',
    yaxis_title='Values',
    template='plotly_white'
)
fig.show()
```

Question7.  Write a program to perform element-wise multiplication of two NumPy arrays.

Question 8.  Create a line plot with multiple lines using Matplotlib.



```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [15, 25, 35, 45, 55]
y3 = [5, 15, 25, 35, 45]

plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--')
plt.plot(x, y3, label='Line 3', color='red', linewidth=2)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Multiple Lines')
plt.legend()

plt.grid(True)
plt.show()
```

Question 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.



```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 22],
    'Salary': [50000, 60000, 70000, 80000, 45000]
}

df = pd.DataFrame(data)

threshold = 30

filtered_df = df[df['Age'] > threshold]

print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (Age > 30):")
print(filtered_df)
```

Question 11.  Perform matrix multiplication using NumPy.

```python
import numpy as np

matrix1 = np.array([[1, 2], [3, 4], [5, 6]])
matrix2 = np.array([[7, 8], [9, 10]])

result = np.matmul(matrix1, matrix2)

print("Matrix 1:")
print(matrix1)
print("\nMatrix 2:")
print(matrix2)
print("\nResult of Matrix Multiplication:")
print(result)
```

Question 13. Create a 3D scatter plot using Plotly.


```python
import plotly.express as px
import pandas as pd

data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 20, 30, 40, 50],
    'Z': [5, 15, 25, 35, 45],
    'Color': ['red', 'blue', 'green', 'purple', 'orange']
}
df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Color', title="3D Scatter Plot")

fig.show()
```
